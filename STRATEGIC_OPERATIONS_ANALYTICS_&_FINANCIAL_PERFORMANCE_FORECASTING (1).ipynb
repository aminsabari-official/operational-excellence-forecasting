```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Membuat data simulasi berdasarkan pencapaian di CV Anda
# Mencerminkan kenaikan omzet 20% dan efisiensi biaya 10%
data = {
    'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun'],
    'Omzet_Juta': [100, 110, 125, 120, 135, 150], # Tren naik (Total +20%+)
    'Biaya_Operasional': [40, 39, 38, 37, 36, 35]   # Tren turun (Efisiensi ~10%)
}

df = pd.DataFrame(data)

# 2. Menghitung Profit
df['Profit'] = df['Omzet_Juta'] - df['Biaya_Operasional']

# 3. Visualisasi Tren (Menggunakan gaya profesional)
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Plot Omzet vs Profit
plt.plot(df['Bulan'], df['Omzet_Juta'], marker='o', label='Omzet (Juta)', color='blue', linewidth=2)
plt.plot(df['Bulan'], df['Profit'], marker='s', label='Net Profit', color='green', linestyle='--')

# Anotasi Pencapaian Manajerial (Sesuai CV Amin Sabari)
plt.annotate('Target Efisiensi 10% Tercapai', xy=(5, 35), xytext=(3, 60),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Performance Report: Revenue Growth & Cost Efficiency', fontsize=15)
plt.xlabel('Periode (Bulan)', fontsize=12)
plt.ylabel('Nilai (Juta Rupiah)', fontsize=12)
plt.legend()
plt.show()

# 4. Ringkasan Executive
print("--- Executive Summary ---")
growth = ((df['Omzet_Juta'].iloc[-1] - df['Omzet_Juta'].iloc[0]) / df['Omzet_Juta'].iloc[0]) * 100
efficiency = ((df['Biaya_Operasional'].iloc[0] - df['Biaya_Operasional'].iloc[-1]) / df['Biaya_Operasional'].iloc[0]) * 100

print(f"Total Revenue Growth: {growth:.1f}%")
print(f"Operational Cost Efficiency: {efficiency:.1f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# 1. SETUP DATA: Berdasarkan skala kepemimpinan 50+ staff & efisiensi 10%
# Data mencakup performa 6 bulan terakhir di posisi Area Manager
data = {
    'Bulan': [1, 2, 3, 4, 5, 6],
    'Omzet_Juta': [450, 480, 520, 510, 545, 590], # Pertumbuhan signifikan ~20%+
    'Biaya_Ops_Juta': [180, 175, 172, 168, 165, 162], # Penurunan biaya ~10%
    'Jumlah_Staff': [52, 52, 51, 50, 50, 50] # Efisiensi jumlah staff
}

df = pd.DataFrame(data)

# 2. FEATURE ENGINEERING (Menghitung Metrik Bisnis Utama)
df['Net_Profit'] = df['Omzet_Juta'] - df['Biaya_Ops_Juta']
df['Productivity_per_Staff'] = df['Omzet_Juta'] / df['Jumlah_Staff']

# 3. PREDIKSI (FORECASTING): Menggunakan Linear Regression
# Memprediksi Omzet untuk bulan ke-7
X = df[['Bulan']]
y = df['Omzet_Juta']
model = LinearRegression().fit(X, y)
prediksi_bulan_7 = model.predict([[7]])[0]

# 4. VISUALISASI DASHBOARD PROFESIONAL
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))
sns.set_style("darkgrid")

# Graph 1: Revenue vs Cost Efficiency
ax1.plot(df['Bulan'], df['Omzet_Juta'], marker='o', color='b', linewidth=3, label='Revenue')
ax1.fill_between(df['Bulan'], df['Omzet_Juta'], color='blue', alpha=0.1)
ax1.plot(df['Bulan'], df['Biaya_Ops_Juta'], marker='x', color='r', linestyle='--', label='Ops Cost')
ax1.set_title('Strategic Performance: Revenue Growth vs Cost Efficiency', fontsize=14)
ax1.legend()

# Graph 2: Staff Productivity Trend
sns.barplot(x='Bulan', y='Productivity_per_Staff', data=df, ax=ax2, palette='viridis')
ax2.set_title('Operational Excellence: Staff Productivity (Revenue/Head)', fontsize=14)

plt.tight_layout()
plt.show()

print("="*60)
print("      EXECUTIVE SUMMARY REPORT - AREA MANAGEMENT")
print("="*60)
print(f"1. Overall Revenue Growth   : {((df['Omzet_Juta'].iloc[-1]/df['Omzet_Juta'].iloc[0])-1)*100:.2f}%")
print(f"2. Cost Efficiency Achieved : {((1 - df['Biaya_Ops_Juta'].iloc[-1]/df['Biaya_Ops_Juta'].iloc[0]))*100:.2f}%")
print(f"3. Staff Productivity Incr. : {((df['Productivity_per_Staff'].iloc[-1]/df['Productivity_per_Staff'].iloc[0])-1)*100:.2f}%")
print(f"4. Next Month Projection    : Rp {prediksi_bulan_7:.2f} Million")
print("="*60)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# 1. DATA INGESTION (Simulasi data berdasarkan CV Amin Sabari)
# Menggambarkan efisiensi 10% dan optimasi workflow di PT Wahana Ottomitra Multiartha
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Efficiency_Index': [80, 82, 85, 88, 92, 95],          # Workflow Optimization
    'Operational_Costs': [200, 195, 190, 185, 182, 180],   # Cost Reduction (10%)
    'Data_Integrity_Rate': [95, 96, 98, 99, 100, 100],     # 100% Data Integrity
    'Monthly_Revenue': [500, 520, 550, 580, 610, 650]      # Scalability Growth
}

df = pd.DataFrame(data)
df['Month_Num'] = np.arange(1, 7)

# 2. PREDICTIVE MODELING (Forecasting Revenue untuk Bulan ke-7)
model = LinearRegression()
X = df[['Month_Num']]
y = df['Monthly_Revenue']
model.fit(X, y)
next_month_pred = model.predict([[7]])[0]

# 3. VISUALIZATION DASHBOARD
plt.figure(figsize=(16, 10))
sns.set_style("whitegrid")

# Subplot 1: Efficiency vs Cost (Korelasi Strategis)
plt.subplot(2, 2, 1)
plt.plot(df['Month'], df['Efficiency_Index'], marker='o', label='Efficiency Index (%)', color='green', linewidth=3)
plt.plot(df['Month'], df['Operational_Costs'], marker='s', label='Ops Cost (Juta)', color='red', linestyle='--')
plt.title('Impact of Workflow Automation on Costs', fontsize=12)
plt.legend()

# Subplot 2: Data Integrity Trend (Target 100%)
plt.subplot(2, 2, 2)
sns.barplot(x='Month', y='Data_Integrity_Rate', data=df, palette='Blues_d')
plt.axhline(100, color='gold', linestyle='--', label='Target: 100% Integrity')
plt.title('Data Integrity & Error Elimination Rate', fontsize=12)
plt.ylim(90, 105)

# Subplot 3: Revenue Growth & Prediction
plt.subplot(2, 1, 2)
plt.plot(df['Month'], df['Monthly_Revenue'], marker='D', label='Actual Revenue', color='blue', linewidth=2)
plt.scatter(['Jul'], [next_month_pred], color='orange', s=150, label='Predicted Revenue (Jul)', zorder=5)
plt.title('Business Scalability & Predictive Growth Forecast', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()

# 4. (Executive Summary)
print("="*70)
print(f"{'STRATEGIC OPERATIONS REPORT - AMIN SABARI':^70}")
print("="*70)
efficiency_gain = df['Efficiency_Index'].iloc[-1] - df['Efficiency_Index'].iloc[0]
cost_reduction = ((df['Operational_Costs'].iloc[0] - df['Operational_Costs'].iloc[-1]) / df['Operational_Costs'].iloc[0]) * 100

print(f"1. Peningkatan Efisiensi Workflow : +{efficiency_gain}% (Melalui Otomatisasi)")
print(f"2. Reduksi Biaya Operasional     : {cost_reduction:.1f}% (Sesuai Target CV)")
print(f"3. Status Integritas Data        : {df['Data_Integrity_Rate'].iloc[-1]}% (Zero Human Error)")
print(f"4. Proyeksi Omzet Bulan Depan    : Rp {next_month_pred:.2f} Juta")
print("="*70)
```
